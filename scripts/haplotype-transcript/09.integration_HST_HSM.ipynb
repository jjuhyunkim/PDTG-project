# Integrative analysis with Haplotype specific transcriptome and Haplotype specific Methylation

## Make Env

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
from scipy.stats import pearsonr
```

```python
# input data
noiseq_result = "/vf/users/Phillippy/projects/HG002_Masseq/06.pantranscriptome/07.hg002v1.0.1/05.quantification/noiseq.result.txt"
hsm_result = "/vf/users/Phillippy/projects/HG002_Masseq/10.interpretation/HST-HSM/HSM.tsv"
integration_result = "/vf/users/Phillippy/projects/HG002_Masseq/10.interpretation/HST-HSM/HST.padding1000.intersect_HSM.tsv"
```

```python
mat = pd.read_csv("/vf/users/Phillippy/projects/HG002_Masseq/10.interpretation/HST-HSM/HSM.mat.coor", sep ='\t' , header = 0)
pat = pd.read_csv("/vf/users/Phillippy/projects/HG002_Masseq/10.interpretation/HST-HSM/HSM.pat.coor", sep ='\t' , header = 0)
```

```python
trans_gene = pd.read_csv("/data/Phillippy/projects/HG002_Masseq/03.align/HG002_Bulk/01.mapped/chm13_2312_2/gtf/chm13v2.0.refseq_curated5.1.rCRS.chrEBV.combined.gene-geneName.pair",
                        sep='\t', header=None)
inc_nov = pd.read_csv("/data/Phillippy/projects/HG002_Masseq/03.align/HG002_Bulk/01.mapped/chm13_2312_2/gtf/chm13v2.0.refseq_curated5.1.rCRS.chrEBV.combined.transcript-gene.includeNoval.pair", 
                      sep='\t', header=None)
trans_gene.columns = ['gene','geneName']
inc_nov.columns = ['isoform','gene']
test = trans_gene.merge(inc_nov, on='gene', how='inner')
test.head()
```

```python
hsm_coor = mat.merge(pat, on = 'hsm_region' , how = 'outer')
```

```python
hst = pd.read_csv(noiseq_result, sep = '\t')
hst.columns = ["hst_" + col for col in hst.columns.tolist()]
hst['isoform'] = hst.index.tolist()
hst.head(2)
```

```python
hsm_hsm = pd.read_csv(integration_result, sep = '\t', header = None)

hsmcol = ["chr","start","end","length","nCG","meanMethy1","meanMethy2","diff.Methy","areaStat",'region']
hsmcol = ["hsm_" + col for col in hsmcol]

isocol = ['chr','start','end']
isocol = ["hst_" + col for col in isocol]

hsm_hsm.columns = hsmcol + isocol + ['isoform','genename','tss']
hsm_hsm = hsm_hsm.merge(hsm_coor, on = "hsm_region", how = 'left')
hsm_hsm.head(5)
```

```python
unique_isoforms = hsm_hsm['isoform'].value_counts() == 1
hsm_hsm['uniqMet'] = ''
hsm_hsm.loc[hsm_hsm['isoform'].isin(unique_isoforms[unique_isoforms].index), 'uniqMet'] = 'uniq_met'
```

```python
hsm_hsm.head(2)
```

## Join datasets

```python
dat = pd.merge(hsm_hsm,hst, on = 'isoform', how = "inner")
dat.head()
```

```python
dat['log2_hsm_areaStat'] = np.log10(abs(dat['hsm_areaStat']))
# Update values in column "b" where the condition is true
dat.loc[dat['hsm_areaStat'] < 0, 'log2_hsm_areaStat'] *= -1
dat.head(2)
```

```python
max(dat['hsm_length'])
```

```python
dat['hsm_middle'] = (dat['hsm_start'] + dat['hsm_end']) / 2
dat['hsm-tss_dis'] = abs(dat['hsm_middle']-dat['tss'])
```

```python
dat.sort_values(by="hsm-tss_dis")[['hsm_length','hsm_middle','tss','hsm-tss_dis']].head()
```

```python
plt.hist(abs(dat['hsm_middle']-dat['tss']), bins=50, color='skyblue', edgecolor='black')
plt.xlabel('distance')
plt.ylabel('Frequency')
plt.title('Distribution of dist between tss and DMR')
plt.show()
```

```python
dat.shape
```

```python
x_values = dat['hsm_areaStat']
y_values = dat['hst_M']
col = dat['uniqMet']
# Create the scatter plot
plt.scatter(x_values, y_values,
           c=col.map({'':'blue','uniq_met':'red'}))

# Add labels and title
plt.xlabel('mathlation diff:mat/pat')
plt.ylabel('expression diff:mat/pat')
plt.title('Integrative plot of difference between mat and pat')

# Show the plot
plt.show()
```

```python
dat['hsm_areaStat'].corr(dat['hst_M'])
```

```python
correlation, p_value = pearsonr(dat['hsm_areaStat'], dat['hst_M'])
```

```python
dat = dat.merge(test, on = 'isoform', how = 'left')
dat = dat.drop_duplicates()
```

```python
dat['sig'] = dat['hsm_areaStat'] * dat['hst_M']
dat = dat.sort_values(by='sig')
```
