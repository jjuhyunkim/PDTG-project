# load data

```r
rpvg = read.table("/data/Phillippy/projects/HG002_Masseq/06.pantranscriptome/07.hg002v1.0.1/05.quantification/hg002v101.allGene.isoquantModel.rpvg_joint.max.txt",
                 row.names = 1,
                 header = TRUE,
                 sep = '\t')
```

```r
head(rpvg)
```

```r
df <- rpvg[c('ReadCount_1','ReadCount_2')]
numeric_cols <- sapply(df, is.numeric)
df[numeric_cols] <- lapply(df[numeric_cols], function(x) as.integer(round(as.numeric(x))))
df <- matrix(c(df$ReadCount_1,df$ReadCount_2),ncol=2)
```

```r
colnames(df) <- c("mat",'pat')
row.names(df) <- c(rpvg$trans)
head(df)
```

```r
fac <- as.data.frame(c('mat','pat'))
row.names(fac) <- c('mat','pat')
colnames(fac) <- "hap"
head(fac)
```

```r
mydata <- readData(data = df, factors = fac)
result <- noiseq(mydata, factor = "hap", k = NULL, norm = "n", pnr = .2, nss = 5, v= 0.02, lc=1, replicates = "no")
result_mat <- as.data.frame(result@results)
```

```r
result_mat['rank_abs'] <- abs(result_mat$ranking)
```
