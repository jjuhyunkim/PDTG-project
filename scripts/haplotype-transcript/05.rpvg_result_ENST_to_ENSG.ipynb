```r
tab['ensembl_transcript_id'] = tab['trans']
```

```r
ensembl <- useMart("ensembl", dataset = "hsapiens_gene_ensembl")
attributes = c("ensembl_transcript_id", "ensembl_gene_id")
```

```r
transcript_ids = tab['trans']
```

```r
gene_ids = getBM(attributes = attributes, filters = "ensembl_transcript_id", 
                 values = transcript_ids, mart = ensembl)
```

```r
gene_ids = data.frame(gene_ids)
```

```r
tab = merge(tab, gene_ids, all.x = TRUE, by = 'ensembl_transcript_id')
```

```r
tab %>% filter(ensembl_gene_id == "ENSG00000141956")
```

```r
tab['totalTPM'] = tab['TPM_1'] + tab['TPM_2'] + 1
tab['totalTPM_log'] = log10(tab['totalTPM'])
```

```r
head(tab,2)
```

```r
sub = tab %>% filter(ensembl_gene_id == "ENSG00000141956")
```

```r
ggplot(sub, aes(y=totalTPM_log, x=trans)) + 
    geom_bar(position="dodge", stat="identity") +
    coord_flip()
```

```r
sub
```

```r
prob_1 =  sub[c("Name_1","TPM_1","totalTPM","ensembl_gene_id","trans")]
prob_2 =  sub[c("Name_2","TPM_2","totalTPM","ensembl_gene_id","trans")]
```

```r
colnames(prob_1) = c("Name","TPM","totalTPM","ensembl_gene_id","trans")
colnames(prob_2) = c("Name","TPM","totalTPM","ensembl_gene_id","trans")
```

```r
prob_1['hap'] = 'hap1'
prob_2['hap'] = 'hap2'
```

```r
prob = rbind(prob_1, prob_2)
```

```r
prob['prob'] = prob['TPM'] / prob['totalTPM']
```

```r
head(prob)
```

```r
ggplot(prob, aes(fill=hap, y=prob, x=trans)) + 
    geom_bar(position="dodge", stat="identity") +
    coord_flip()
```
