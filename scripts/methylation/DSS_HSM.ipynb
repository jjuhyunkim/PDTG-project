```r
suppressPackageStartupMessages(library(DSS))
suppressPackageStartupMessages(library(bsseq))
options(warn=-1)
#options(warn=0) # use to turn warining back on.
```

```r
chr=22
outDir="/data/Phillippy/projects/HG002_Masseq/09.methylation/steven_count"
```

```r
dat1.1 = paste0("/vf/users/Phillippy/projects/HG002_Masseq/09.methylation/steven_count/steven.ont.hg002.mat.toCHM13.chr",as.character(chr),".dmr.bed")
dat1.2 = paste0("/vf/users/Phillippy/projects/HG002_Masseq/09.methylation/steven_count/steven.ont.hg002.pat.toCHM13.chr",as.character(chr),".dmr.bed")
dat2.1 = paste0("/vf/users/Phillippy/projects/HG002_Masseq/09.methylation/steven_count/steven.hifi.hg002.mat.toCHM13.chr",as.character(chr),".dmr.bed")
dat2.2 = paste0("/vf/users/Phillippy/projects/HG002_Masseq/09.methylation/steven_count/steven.hifi.hg002.pat.toCHM13.chr",as.character(chr),".dmr.bed")
```

```r
dat1.1 = read.table(file.path(dat1.1), header=TRUE)
dat1.2 = read.table(file.path(dat1.2), header=TRUE)
dat2.1 = read.table(file.path(dat2.1), header=TRUE)
dat2.2 = read.table(file.path(dat2.2), header=TRUE)
```

```r
BSobj = makeBSseqData( list(dat1.1, dat1.2, dat2.1 , dat2.2),
      c('mat_ont','pat_ont','mat_hifi','pat_hifi'))
BSobj
```

```r
dmlTest.sm = DMLtest(BSobj, group1=c("mat_ont","mat_hifi"), group2=c("pat_ont","pat_hifi"), 
                     smoothing=TRUE, ncores = 20)
```

```r
dmls = callDML(dmlTest.sm)
# dmls2 = callDML(dmlTest, delta=0.1, p.threshold=0.001)
head(dmls)
```

```r
dmrs = callDMR(dmlTest.sm)
head(dmrs)
```

```r
dmrs = callDMR(dmlTest.sm)
head(dmrs)
```

```r
showOneDMR(dmrs[1,], BSobj)
```

```r
write.table(dmrs, paste0(outDir, "/dmr.hifi_ont.hap.chr",chr,".tsv"),
           quote = FALSE,
           row.names = FALSE,
           col.names = TRUE,
           sep="\t")
```
