```r
suppressPackageStartupMessages(library(DSS))
suppressPackageStartupMessages(library(bsseq))
options(warn=-1)
#options(warn=0) # use to turn warining back on.
```

```r
t21 = "/home/kimj75/21T.chr22.txt"
n21 = "/home/kimj75/21N.chr22.txt"
t23 = "/home/kimj75/23T.chr22.txt"
n23 = "/home/kimj75/23N.chr22.txt"
```

```r
dat1.1 = read.table(file.path(t21), header=TRUE)
dat1.2 = read.table(file.path(n21), header=TRUE)
dat2.1 = read.table(file.path(t23), header=TRUE)
dat2.2 = read.table(file.path(n23), header=TRUE)
```

```r
dat1.1
```

```r
BSobj = makeBSseqData( list(dat1.1, dat1.2, dat2.1 , dat2.2),
      c('21t','21n','23t','23n')) [200000:500000,]
BSobj
```

```r
dmlTest.sm = DMLtest(BSobj, group1=c("21t","23t"), group2=c("21n","23n"), 
                     smoothing=TRUE, ncores = 1)
```

```r
dmls = callDML(dmlTest.sm , p.threshold=0.001)
# dmls2 = callDML(dmlTest, delta=0.1, p.threshold=0.001)
head(dmls)
```

```r
dmrs = callDMR(dmlTest.sm, p.threshold=0.001)
head(dmrs)
```

```r
showOneDMR(dmrs[1,], BSobj)
```
